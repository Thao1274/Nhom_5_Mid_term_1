```python
# Doanh thu theo tỉnh
province_revenue = df.groupby("Province")["TicketPrice"].sum().reset_index().sort_values(by="TicketPrice", ascending=False)
top_n = 4  
top_provinces = province_revenue.sort_values(ascending=False).head(top_n)
other = province_revenue.drop(top_provinces.index).sum()
province_plot = top_provinces.append(pd.Series({"Khác": other}))


plt.figure(figsize=(7, 7))
plt.pie(province_plot, labels=province_plot.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette("Pastel1"))
plt.title("Tỷ trọng doanh thu theo tỉnh")
plt.axis("equal")
plt.tight_layout()
plt.show()
```

```python

# Doanh thu theo thể loại phim chính
df_explode = df.copy()
df_explode["GenreList"] = df_explode["Genres"].str.split(",")
df_explode = df_explode.explode("GenreList")
df_explode["GenreList"] = df_explode["GenreList"].str.strip()
genre_exploded_revenue = df_explode.groupby("GenreList")["TicketPrice"].sum().reset_index()
genre_exploded_revenue = genre_exploded_revenue.sort_values(by="TicketPrice", ascending=False)
plt.figure(figsize=(10, 6))
sns.barplot(data=genre_exploded_revenue.head(5), x="TicketPrice", y="GenreList", palette="coolwarm")
plt.title("Doanh thu theo từng thể loại phim (Top 5)")
plt.xlabel("Doanh thu (VNĐ)")
plt.ylabel("Thể loại")
plt.tight_layout()
plt.show()
```

```python

# Tỷ lệ khách mua bỏng ngô
popcorn_ratio = df["Popcorn"].value_counts(normalize=True)
plt.figure(figsize=(5, 5))
popcorn_ratio.plot(kind='pie', autopct='%1.1f%%', labels=popcorn_ratio.index, colors=["#FFD700", "#C0C0C0"])
plt.title("Tỷ lệ khách mua bỏng ngô")
plt.ylabel("")
plt.tight_layout()
plt.show()
```

```python
# Doanh thu theo các thứ trong tuần

df["Weekday"] = df["SaleDate"].dt.day_name()  # ví dụ: Monday, Tuesday, ...


weekday_revenue = df.groupby("Weekday")["TicketPrice"].sum().reset_index()


weekday_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
weekday_revenue["Weekday"] = pd.Categorical(weekday_revenue["Weekday"], categories=weekday_order, ordered=True)
weekday_revenue = weekday_revenue.sort_values("Weekday")


plt.figure(figsize=(10, 5))
sns.barplot(data=weekday_revenue, x="Weekday", y="TicketPrice", palette="viridis")
plt.title("Doanh thu theo các thứ trong tuần")
plt.ylabel("Doanh thu (VNĐ)")
plt.xlabel("Thứ ")
plt.tight_layout()
plt.show()
```

```python

# Số lượng vé theo thể loại chỗ

seat_type_counts = df["Thể loại chỗ"].value_counts().reset_index()
seat_type_counts.columns = ["SeatType", "Count"]

# 
plt.figure(figsize=(6, 4))
sns.barplot(data=seat_type_counts, x="SeatType", y="Count", palette="Set1")
plt.title("Số lượng vé theo loại chỗ ngồi (Đơn/Đôi)")
plt.xlabel("Loại chỗ")
plt.ylabel("Số lượng vé")
plt.tight_layout()
plt.show()

```

```python
# Số lượng vé theo thời gian chiếu

df["Giờ"] = df["Thời gian"].str[:2]  # Lấy giờ (2 ký tự đầu)

time_counts = df["Giờ"].value_counts().sort_index()

# 
plt.figure(figsize=(10, 4))
sns.barplot(x=time_counts.index, y=time_counts.values, palette="cubehelix")
plt.title("Số lượng vé theo thời gian chiếu (giờ)")
plt.xlabel("Giờ chiếu (24h)")
plt.ylabel("Số lượng vé")
plt.tight_layout()
plt.show()
```

```python
# Số lượng vé theo độ dài phim
# 
df["Thời lượng phim"] = pd.to_numeric(df["Thời lượng phim"], errors='coerce')

# 
bins = [0, 90, 105, 120, 135, 150, 200]
labels = ["<90p", "90-105p", "105-120p", "120-135p", "135-150p", ">150p"]
df["DurationGroup"] = pd.cut(df["Thời lượng phim"], bins=bins, labels=labels)

# 
duration_counts = df["DurationGroup"].value_counts().sort_index()

# 
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 4))
sns.barplot(x=duration_counts.index, y=duration_counts.values, palette="ch:s=.25,rot=-.25")
plt.title("Số lượng vé theo độ dài phim")
plt.xlabel("Thời lượng phim")
plt.ylabel("Số lượng vé")
plt.tight_layout()
plt.show()
```

```python
# Số lượng vé theo quốc gia phát hành
# 
df = df.rename(columns={"Đất nước": "Country"})

country_counts = df["Country"].value_counts().reset_index()
country_counts.columns = ["Country", "Count"]

# 
plt.figure(figsize=(8, 4))
sns.barplot(data=country_counts.head(6), x="Country", y="Count", palette="cool")
plt.title("Số lượng vé theo quốc gia sản xuất phim (Top 6)")
plt.xlabel("Quốc gia")
plt.ylabel("Số vé")
plt.tight_layout()
plt.show()
```
